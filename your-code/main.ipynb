# Challenge 2 - Data Cleaning and Transformation

```python
# Your code here
# Store spending columns
spending_cols = customers.columns.drop(['Channel', 'Region'])
clean_customers = customers[spending_cols].copy()

# Apply Log Transformation (log(1 + x)) to reduce skewness and outlier impact
clean_customers_log = np.log1p(clean_customers)
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:
from sklearn.preprocessing import StandardScaler

# Your code here:
# Scale the log-transformed data
scaler = StandardScaler()
customers_scale = scaler.fit_transform(clean_customers_log)
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
kmeans_2 = KMeans(n_clusters=2).fit(customers_scale)

labels = kmeans_2.predict(customers_scale)

clusters = kmeans_2.labels_.tolist()
```

```python
clean_customers['Label'] = clusters
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN 

# Initiate and fit the DBSCAN model with eps=0.5 (default min_samples=5)
dbscan = DBSCAN(eps=0.5).fit(customers_scale)

# Assign labels back to the DataFrame
labels_DBSCAN = dbscan.labels_
clean_customers['DBSCAN_Label'] = labels_DBSCAN

```

Count the values in `labels_DBSCAN`.

```python
# Count the values in labels_DBSCAN
clean_customers['DBSCAN_Label'].value_counts()
```

# Challenge 6 - Compare K-Means with DBSCAN

Now we want to visually compare how K-Means and DBSCAN have clustered our data. We will create scatter plots for several columns. For each of the following column pairs, plot a scatter plot using `labels` and another using `labels_DBSCAN`. Put them side by side to compare. Which clustering algorithm makes better sense?

Columns to visualize:

* `Detergents_Paper` as X and `Milk` as y
* `Grocery` as X and `Fresh` as y
* `Frozen` as X and `Delicassen` as y

Visualize `Detergents_Paper` as X and `Milk` as y by `labels` and `labels_DBSCAN` respectively

```python
def plot_comparison(df, x_col, y_col):
    """Generates a comparison of K-Means and DBSCAN labels on raw spending data."""
    plot_df = df.copy()
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # K-Means Plot
    sns.scatterplot(x=x_col, y=y_col, hue='KMeans_Label', data=plot_df, palette='viridis', legend='full', ax=axes[0])
    axes[0].set_title(f'K-Means Clusters ({x_col} vs {y_col})')
    
    # DBSCAN Plot
    sns.scatterplot(x=x_col, y=y_col, hue='DBSCAN_Label', data=plot_df, palette='plasma', legend='full', ax=axes[1])
    axes[1].set_title(f'DBSCAN Clusters ({x_col} vs {y_col})')
    
    plt.tight_layout()
    plt.show()
```

```python
# Your code here:
original_spending = customers.copy()

# 2. Add the KMeans and DBSCAN labels from the clean_customers DataFrame
# Note: We use .index to ensure alignment.
original_spending['KMeans_Label'] = clean_customers['KMeans_Label']
original_spending['DBSCAN_Label'] = clean_customers['DBSCAN_Label']

# --- Then run your plot code (which will now work) ---
plot_comparison(original_spending, 'Detergents_Paper', 'Milk')
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here:

# Example for K=4 (another common finding for this dataset)
kmeans_4 = KMeans(n_clusters=4, random_state=42, n_init=10).fit(customers_scale)
clean_customers['KMeans_Label_4'] = kmeans_4.labels_

# Plot a comparison (e.g., Fresh vs Grocery)
fig, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(x='Fresh', y='Grocery', hue='KMeans_Label_4', data=original_spending.assign(KMeans_Label_4=kmeans_4.labels_), palette='viridis', legend='full')
ax.set_title('K-Means Clusters (k=4) - Fresh vs Grocery')
plt.show()
```

**Your comment here**

- Experimentation (often using the Elbow Method or Silhouette Score on the full dataset) generally shows that k=2 or k=3 are the most interpretable number of clusters for this data, often corresponding to the Channels (k=2) or the Regions (k=3).

- Increasing k (e.g., to 4 or 5) may provide finer segment details, but the marginal gain in interpretability often decreases, and the clusters may still be heavily influenced by the Channel variable.

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python

# Experiment 1: Increase eps to 1.0 (allowing a larger neighborhood)
dbscan_exp1 = DBSCAN(eps=1.0).fit(customers_scale)
clean_customers['DBSCAN_Label_1.0'] = dbscan_exp1.labels_

# Experiment 2: Keep eps=0.5 but decrease min_samples to 2 (allowing smaller, denser clusters)
dbscan_exp2 = DBSCAN(eps=0.5, min_samples=2).fit(customers_scale)
clean_customers['DBSCAN_Label_0.5_min2'] = dbscan_exp2.labels_

print("DBSCAN 1.0 Label Counts:")
print(clean_customers['DBSCAN_Label_1.0'].value_counts())
print("\nDBSCAN 0.5 (min_samples=2) Label Counts:")
print(clean_customers['DBSCAN_Label_0.5_min2'].value_counts())
```
